## Decision Tree Implementation

```python
import pandas as pd
import numpy as np


class Node:
 
    def __init__(self, node_id):
        self.node_id = node_id
        
        self.left = None 
        self.right = None
        self.leaf = None
        self.class_label = None
        
        self.samples : int = None
        self.items : list = None
        self.info : float = None
        self.feature_index = None
        self.threshold = None
        

class DecisionTreeClassifier:
    def __init__(self, max_depth=None, min_samples_split=None, cost_of_feature_file_name=None):
        self.root = None
        self.node_id = 0
        self.fitted = None
        
        # list of classes
        self.classes = None
        # total number of features provided in training set
        self.number_of_features = None
        
                        # cost of feature
        # pandas data frame
        self.cost_of_features = None
        if cost_of_feature_file_name is not None:
            self.cost_of_features = self.extract_cost_of_feature(cost_of_feature_file_name)
            
                        # pre-prunning       
        # max-depth
        self.max_depth = max_depth
        self.cur_depth = -1 # root node will be the 0'th step
        
        # min samples a leaf has
        self.min_samples_split = min_samples_split
        
    
    def extract_cost_of_feature(self, file_name):
        with open(file_name, 'r') as f:
            column_names = []
            values = []
            
            for line in f:
                column_name, value = line.strip().split(':')
                column_name = column_name.strip()
                value = value.strip()
                column_names.append(column_name)
                values.append(value)
            
            column_names.append('21st')
            values.append('1.00') # for updated splitting criterion
            
            df = pd.DataFrame(data={col: [] for col in column_names})
            df = df.append(pd.Series(values, index=df.columns, dtype=float), ignore_index=True)
        return df
    
   
    def get_new_node_id(self):
        self.node_id += 1
        return self.node_id
   
    def set_number_of_classes(self, labels: np.array):
        self.classes = np.unique(labels)
    
    
    def set_number_of_features(self, data):
        # data is pandas dataframe 
        self.number_of_features = len(data.columns)


    def is_leaf_pure(self, labels) -> bool:
        return np.unique(labels).size == 1


    def count_classes(self, labels):
        max_ = max(self.classes)
        min_ = min(self.classes)
        
        occ = [0 for _ in range(max_+1)]
        
        uniques, counts = np.unique(labels, return_counts=True)
        for uniq, count in zip(uniques, counts):
            occ[uniq] = count
                
        return  occ[min_:max_+1] 


    def feature_threshold_picker(self, col, labels):
        
        col = col.values
        labels = labels.values

        sorted_indices = np.argsort(col)
        sorted_col = col[sorted_indices]
        sorted_labels = labels[sorted_indices]

        threshold = None

        # binary search for peak finding, unimodal array, it has a single peak
        low = 0
        high = len(col) - 1
        mid = None

        while low <= high:
            mid = low + (high-low)//2
            _, _, _, left_gain = self.gini(sorted_labels, sorted_labels[:mid], sorted_labels[mid:])
            _, _, _, mid_gain = self.gini(sorted_labels, sorted_labels[:mid+1], sorted_labels[mid+1:])
            _, _, _, right_gain = self.gini(sorted_labels, sorted_labels[:mid+2], sorted_labels[mid+2:])
            if  left_gain < mid_gain and right_gain <= mid_gain:
                break
            elif mid_gain <= right_gain:
                # move right
                low = mid + 1
            elif mid_gain <= left_gain:
                # move left
                high = mid - 1
        
        # to select rightmost index in the repeating values
        tmp = mid + 1
        while (tmp <len(col) 
               and np.isclose(col[mid], col[tmp], 1e-8)
              ):
            mid += 1
            tmp += 1

        if mid != (len(col)-1):
            threshold = (sorted_col[mid] + sorted_col[mid+1]) / 2
        else:
            threshold = sorted_col[mid]
        assert threshold is not None
        return threshold


    def partition(self, feature_col, labels):
        left_indices = list()
        right_indices = list()

        threshold = None

        if feature_col.dtype == 'int64':  # for integers
            # print("1-0 here")
            # partition
            threshold = 0.5
            for idx, val in enumerate(feature_col, start=0):
                if val == 0:
                    left_indices.append(idx)
                elif val == 1:
                    right_indices.append(idx)

        elif feature_col.dtype == 'float64':  # for floats
            # print("float here")

            # pick the best threshold
            threshold = self.feature_threshold_picker(feature_col, labels)

            # partition
            for idx, val in enumerate(feature_col, start=0):
                if val < threshold:
                    left_indices.append(idx)
                else:
                    right_indices.append(idx)
        else:
            assert "error"

        return left_indices, right_indices, threshold

    
    def calc_gini(self, labels):
        gini_val = 1
        total = len(labels)
        counts = self.count_classes(labels)
        for count in counts:
            if count != 0:
                gini_val -= (count/total)**2
        return gini_val

    
    def gini(self, labels, ll, rl):
        parent_info = self.calc_gini(labels)
        left_info = self.calc_gini(ll)
        right_info = self.calc_gini(rl)
        child_info = len(ll)/len(labels)*left_info + len(rl)/len(labels)*right_info
        gain = parent_info - child_info
        return parent_info, left_info, right_info, gain

    
    def calc_best_split(self, data, labels):

        # data will be a dataframe
        max_gain = float('-inf')
        info = None

        left_indices = list()
        right_indices = list()
        max_feature_index = -1
        max_feature_threshold = None
        
        for feature_index in range(self.number_of_features):
            feature_col = data.iloc[:,feature_index]

            gini_list = list()
            left_data = list()
            right_data = list()

            # partition. returns list of indices
            left_idxs, right_idxs, threshold = self.partition(feature_col, labels)

            # create new dataframes according to partition
            #ld: left data, ll: left label, rd: right data, rl: right label
            ld, ll = feature_col.iloc[left_idxs], labels.iloc[left_idxs]
            rd, rl = feature_col.iloc[right_idxs], labels.iloc[right_idxs]

            # calculates gini
            parent_info, left_info, right_info, gain = self.gini(labels, ll, rl)
            
            update_flag = None
            
            # with cost of feature 'Tan 93'
            if self.cost_of_features is not None:
                cost = self.cost_of_features.values[0][feature_index]
                gain = gain**2 / cost
               
            if max_gain <= gain:
                max_gain = gain
                max_feature_index = feature_index
                max_feature_threshold = threshold
                left_indices = left_idxs
                right_indices = right_idxs
                
        return max_feature_index, max_feature_threshold, parent_info, left_indices, right_indices


    def grow_tree(self, data, labels, depth):
        
        def helper_create_leaf_node(labels):
            # initialize as if it is a pure leaf
            node = Node(self.get_new_node_id())
            node.info = 0.0
            node.items = self.count_classes(labels)
            node.samples = len(labels)
            return node
        
        def set_leaf_node(node):
            node.leaf = True
            arr = node.items
            max_val = max(arr)
            max_index = arr.index(max_val)
            node.class_label = max_index + 1  # why +1? labels starts from 1

        # initialize the node that will be returned
        node = helper_create_leaf_node(labels)
        
        if len(labels) == 0:
            return None
    
        # pre-prunning: max depth
        if self.max_depth is not None:
            if depth == self.max_depth:
                set_leaf_node(node)
                return node
        
        # pre-prunning: min samples in the leaf node
        if self.min_samples_split is not None:
            if self.min_samples_split >= len(labels):
                set_leaf_node(node)
                return node
            
        # stopping criteria: pure node
        if self.is_leaf_pure(labels):
            set_leaf_node(node)
            return node

        # left is true part, right is the remaning part of the data
        feature_index, threshold, parent_info, l_idxs, r_idxs = self.calc_best_split(data, labels)
        
        # ! depth is increased manually as an argument to grow_tree method
        left = self.grow_tree(data.iloc[l_idxs], labels.iloc[l_idxs], depth+1)
        right = self.grow_tree(data.iloc[r_idxs], labels.iloc[r_idxs], depth+1)

        node.info = parent_info  # gini value
        node.items = self.count_classes(labels)
        node.samples = len(labels)
        node.threshold = threshold
        node.left = left
        node.right = right
        node.feature_index = feature_index

        return node

    
    def fit(self, data, labels):
        self.set_number_of_classes(labels)
        self.set_number_of_features(data)
        self.root = self.grow_tree(data, labels, 0)
        self.fitted = True
        
        
    def predict(self, data):
        assert self.fitted is not None, "Decision Tree not fitted"
        
        preds = pd.Series(dtype='int64')  # class labels are integers
        
        for index, row in data.iterrows():       
            cur_node = self.root
            while cur_node:
                if cur_node.leaf is not None:
                    preds = pd.concat([preds, pd.Series(cur_node.class_label, index=[index])])
                    break
                else:
                    f_idx = cur_node.feature_index
                    value = row.iloc[f_idx]
    
                    if value < cur_node.threshold:
                        cur_node = cur_node.left
                    else:
                        cur_node = cur_node.right
        return preds
    
    
    def calc_costs_for_actual_class(self, data):
        assert self.fitted is not None, "Decision Tree not fitted"
        
        preds = pd.Series(dtype='int64')  # class labels are integers
        running_cost = 0
        
        for index, row in data.iterrows():       
            cost = 0
            visited = []
            
            f19, f20, f21 = 18, 19, 20  # index numbers of the feature 19, 20, 21
            
            cost_19 = self.cost_of_features.values[0][f19] 
            cost_20 = + self.cost_of_features.values[0][f20]
            cost_19_20 = cost_19 + cost_20
            
            cur_node = self.root
            while cur_node:
                if cur_node.leaf is not None:
                    preds = pd.concat([preds, pd.Series(cur_node.class_label, index=[index])])
                    break
                else:
                    f_idx = cur_node.feature_index
                    value = row.iloc[f_idx]
                    
                    # calc cost
                    if f_idx != f21:  # last feature
                        cost += self.cost_of_features.values[0][f_idx]
                        if f_idx == f19 or f_idx == f20: 
                            visited.append(f_idx)
                    else:
                        # calculating the cost of unextracted feature
                        tmp_cost = cost_19_20
                        
                        if (f19 in visited):
                            tmp_cost -= cost_19
                        
                        if (f20 in visited):
                            tmp_cost -= cost_20

                        cost += tmp_cost
                    
                    # next node
                    if value < cur_node.threshold:
                        cur_node = cur_node.left
                    else:
                        cur_node = cur_node.right
                        
            running_cost += cost
            
        return running_cost
    
    def report_cost(self, data, labels):
        df = pd.concat([data, labels], axis=1) # combine x_train and y_train into a single DataFrame
        class_groups = df.groupby(df.columns[-1]) 

        for label, group in class_groups:
            running_cost = 0
            instance_cost = self.calc_costs_for_actual_class(group)
            print(f"------\nclass = {label}, avg_cost = {instance_cost/len(group)}\n")
    
        
                    
    def report(self):
        frontiers = [[self.root]]
        i = 0
        while frontiers:

            frontier = frontiers.pop()
            new_frontier = []
            for node in frontier:
                i += 1
                print(i)
                print("value:", node.items )
                print("samples:", node.samples)
                print("feature index:", node.feature_index)

                print("threshold:", node.threshold)
                print("gini:", node.info)
                print("class_label:", node.class_label)
                print("----\n")
                
                if node.left:
                    new_frontier.append(node.left)
                if node.right:
                    new_frontier.append(node.right)
            
            if new_frontier:
                frontiers.append(new_frontier)
    
    def create_vis(self):
        """
        Automatically creates dot format to visualize the graph with graphviz.
        """
        
        # helper function to check an attribute is None or not
        def foo(s):
            if s is None:
                return ""
            return s

        dot = graphviz.Digraph(comment='Tree')
        dot.attr('node', shape='box', margin='0.03,0.03,0.03,0.03')

        frontiers = [[self.root]]
        while frontiers:

            frontier = frontiers.pop()
            new_frontier = []
            for node in frontier:
                if node is None:
                    continue
                    
                # node label for the visualization
                s = ""
                f_idx = str(foo(node.feature_index))  # value 0 act like None in graphviz ?
                f_th = foo(node.threshold)
                if f_idx:
                    s+= f"x[{f_idx}] < {f_th:.3f}\n"
                s += f"gini = {node.info:.3f}\n"
                s += f"samples = {node.samples}\nvalues = {node.items}"
                if node.class_label:
                    s += f"\nclass = {node.class_label}"

                dot.node(str(node.node_id), s)
                
                # edges list is to store node edges for visualiazation
                edges = []

                if node.left:
                    new_frontier.append(node.left)
                    edges.append(str(node.left.node_id))
                if node.right:
                    new_frontier.append(node.right)
                    edges.append(str(node.right.node_id))

                for t in edges:
                    dot.edge(str(node.node_id), t)

            if new_frontier:
                frontiers.append(new_frontier)
                
        
        return dot 
    
```

```python
def class_accuracies(support: np.array, true_preds: np.array, verbose=True) -> list:
    res = list()
    for idx, (base, pred) in enumerate(zip(support, true_preds), start=1):
        acc = pred / base * 100
        if verbose:
            print(f"Label {idx} Accuracy : {acc} percent.")
        res.append(acc)
    return res

def calc_support(labels, preds) -> np.array:
    # support is the array of the occurences of the each class
    _, _, _, support = precision_recall_fscore_support(
                                             labels, preds, 
                                             average=None)
    return support


def confusion_matrix_custom(clf, labels, preds):
    # make prediction
    p = preds.to_numpy()
    l = labels.to_numpy()
    
    # number of classes
    n = len(clf.classes)
    
    # initialize confusion matrix
    cm = np.zeros((n,n))
    
    # populate the confusion matrix
    for label, prediction in zip(l, p):
        cm[label-1][prediction-1] += 1
    
    return cm
    
def plot_confusion(cm, clf):
    # uses matplotlib
    cm_normalized = cm / sum(sum(cm))
    
    fig, ax = plt.subplots()
    im = ax.imshow(cm_normalized, vmin=0, vmax=0.1)
    
    for i in range(len(cm)):
        for j in range(len(cm[0])):
            ax.text(j, i, int(cm[i][j]), ha='center', va='center', color='white')
    
    ticks = [i for i in range(len(cm))]
    ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    ax.set_xticklabels(list(map(str, clf.classes)))
    ax.set_yticklabels(list(map(str, clf.classes)))
    
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')

    #cbar = ax.figure.colorbar(im, ax=ax)

    ax.set_title('Confusion Matrix')

    # displays the confusion matrix
    plt.show()


def get_avg_acc(clf, data, labels):
    preds = clf.predict(data)
    
    # make prediction
    p = preds.to_numpy()
    l = labels.to_numpy()
    
    # number of classes
    n = len(clf.classes)
    
    # initialize confusion matrix
    cm = np.zeros((n,n))
    
    # populate the confusion matrix
    for label, prediction in zip(l, p):
        cm[label-1][prediction-1] += 1
        
    res = []
    for i in range(n):
        acc = cm[i][i]/ np.sum(cm[i])
        res.append(acc)
        
    return sum(res) / n
    
def print_metrics(d, cm: np.array):
    p, r, f = 0, 0, 0
    acc = 0
    # for each label
    print("Class based precision, recall, fcore results:\n")
    for i in range(len(d.classes)):
        tp = cm[i][i]
        fn = np.sum(cm[i]) - tp  # row - tp
        fp = np.sum(cm[:,i]) - tp  # col - tp

        acc = tp / (fn+tp)
        precision = tp / (tp + fp)
        recall = tp / (tp+fn)
        f_score = 2/(1/precision + 1/recall)

        p += precision
        r += recall
        f += f_score
        print(f"class {i+1} -> precision = {precision:.3f} | recall = {recall:.3f} | f_score = {f_score:.3f}")
    real accuracy
    
    print("\nClass accuracies:\n")
    running_corrects = 0
    for i in range(len(d.classes)):
        avg_c = cm[i][i] / np.sum(cm[i])
        running_corrects +=  cm[i][i]
        print(f"\tclass{str(i+1)}".ljust(15, " "), ":", avg_c)
    
    avg_acc = running_corrects / np.sum(np.sum(cm))
    print("\nSet Accuracy".ljust(15, " "),":" , avg_acc)
    
def plot_res(clf, data, labels):
    preds = clf.predict(data)
    cm = confusion_matrix_custom(clf, labels, preds)
    print_metrics(clf, cm)
    _ = plot_confusion(cm, clf)
```

### Random search parameters

```python
criterion = ["gini", "entropy", "log_loss"]
splitter = ["best", "random"]
max_depth = [None] + [5*i for i in range(1, 10)]
min_samples_split = [i for i in range(2,10)] + [i*5 for i in range(2,5)]

print(criterion, splitter, max_depth, min_samples_split, sep="\n")
```

```python
import random
def get_random_parameters(l: list):
    res = []
    for param in l:
        res.append(random.choice(param))
    return res
    
parameters_list = [criterion, splitter, max_depth, min_samples_split]
best_comb = list()
best_acc = -1

for i in range(100):
    cri, spl, m_d, m_s_s = get_random_parameters(parameters_list)
    
    d_search = tree.DecisionTreeClassifier(criterion=cri, splitter=spl,
                                      max_depth=m_d, min_samples_split=m_s_s)
    d_search.fit(x_train, y_train)
    acc = d_search.score(x_val, y_val)

    if acc > best_acc:
        best_acc = acc
        best_comb = [cri, spl, m_d, m_s_s]

print("\n")
print(best_comb, f"val_acc={best_acc}", sep="\n")
d_search = tree.DecisionTreeClassifier(criterion=best_comb[0], splitter=best_comb[1],
                                      max_depth=best_comb[2], min_samples_split=best_comb[3])
d_search.fit(x_train, y_train)
print(d_search.score(x_val, y_val))
print(d_search.score(x_test, y_test))
```

# Confusion Matrix

source: sk-learn  

Can be interpreted as:
- **Presicion:** Ability of the classifier not to label a negative sample as positive.
- **Recall:** Ability of the classifier to find all the positive samples.
- **F-beta score:** Weighted harmonic mean of the presicion and recall, where and F-beta score reaches its best value at 1 and worst score at 0. The F-beta score weights recall moret than precision by a factor of *beta*. *beta = 1.0" means recall and precision are equally important.

**tp:** true positive, **fn:** false negative, **fp:** false positive  
$$precision = \frac{tp}{(tp + fp)} $$  
$$recall = \frac{tp}{(tp + fn)} $$  
$$F1 \;score = \frac{2*precision*recall}{precision+recall}$$

$$F1 \;score = \frac{2*tp}{2 * tp + fp + fn}$$

$$ F-beta \;score = (1 + \beta^{2}) \cdot \frac{precision*recall}{\beta^{2} \cdot precision+recall} $$

---
Geometric representations of the mean calculations:
$$ Quadratic \; Mean \; (RMS) : \sqrt{\frac{a^{2}+b^{2}}{2}}$$  
$$ Arithmetic \; Mean \; : \frac{a+b}{2}$$  
$$ Geometric \; Mean \; : \sqrt{ab}$$  
$$ Harmonic \; Mean \; : \frac{2ab}{a+b}, \; \frac{n}{\frac{1}{a} + \frac{1}{b}} \;\; \text{(set of n numbers \ reciprocal summation)}$$ 


open ai explanation:  
"""Intuitively, the harmonic mean is useful when dealing with rates or ratios that have a reciprocal relationship, such as speed and time. For example, if you travel at two different speeds for the same amount of time, the harmonic mean can be used to calculate your average speed over the entire journey.

The harmonic mean gives more weight to smaller values in the set, which makes it useful in situations where you want to emphasize the importance of smaller values. For example, if you are calculating the average of a set of rates, such as the average number of miles per gallon for a car, the harmonic mean would give more weight to the lower values, which can be useful when comparing fuel efficiency across different types of vehicles."""

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
def class_accuracies(support: np.array, true_preds: np.array, verbose=True) -> list:
    res = list()
    for idx, (base, pred) in enumerate(zip(support, true_preds), start=1):
        acc = pred / base
        if verbose:
            print(f"Label {idx} Accuracy : {acc}")
        res.append(acc)
    return res

def calc_support(labels, preds) -> np.array:
    # support is the array of the occurences of the each class
    _, _, _, support = precision_recall_fscore_support(
                                             labels, preds, 
                                             average=None)
    return support

        
def plot_res(clf, data, labels):
    preds = clf.predict(data)
    cm = confusion_matrix(labels, preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                                  display_labels=clf.classes_)
    
    # 'macro' calculates the unweighted mean for each metric
    presicion, recall, f1score, _  = precision_recall_fscore_support(
                                                        labels, preds, 
                                                        average='macro')
    
    # total number of occurence of each class
    support = calc_support(labels, preds)
    
    # class true predictions / total number of occurences of the class
    accs = accuracy_score(labels, preds)
    
    print(f"{presicion = }\n{recall = }\n{f1score = }\n\n{support = }\n")
    
    # diagonal of the confusion matrix is the true positives for each class (correct predictions)
    true_preds = np.diag(cm)
    _ = class_accuracies(support, true_preds)
    
    set_acc = clf.score(data, labels)
    print("\nSet Accuracy:    ", set_acc)
    
    # displays the confusion matrix
    _ = disp.plot()
```

# c) Pruning the Model

```python
d_prun = tree.DecisionTreeClassifier(criterion='gini', splitter='best')
d_prun.fit(x_train, y_train) 
print("node count:", d_prun.tree_.node_count, "max_depth:", d_prun.tree_.max_depth)
```

```python
path = d_prun.cost_complexity_pruning_path(x_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

fig, ax = plt.subplots()
ax.plot(ccp_alphas[:-1], impurities[:-1], marker="o", drawstyle="steps-post")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
```

## plot accuracy graph

```python
def apply_pruning(ccp_alphas, X, Y):
    clfs = []
    for ccp_alpha in ccp_alphas:
        clf = tree.DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
        clf.fit(X, Y)
        clfs.append(clf)
    return clfs

def calc_accs(clfs, X, Y):
    accs = []
    
    for clf in clfs:
        # calculate class accuracies for xy pairs
        y_preds = clf.predict(X)
        cm = confusion_matrix(Y, y_preds)
        true_preds = np.diag(cm)
        support = calc_support(Y, y_preds)
        acc = class_accuracies(support, true_preds, verbose=False)
        accs.append(acc)
        
    return support, accs
```

```python
def pruning_res_df(clfs, ccp_alphas, X, Y, df_name: str):
    df = pd.DataFrame(columns=["Number of Nodes", "Max Depth", "alpha", "c.1", "c.2", "c.3"])
    df.name = df_name
    
    _, accs = calc_accs(clfs, X, Y)

    for acc, clf, alpha in zip(accs, clfs, ccp_alphas):
        row = [clf.tree_.node_count, clf.tree_.max_depth, alpha]
        acc = ['{:.2f}'.format(num) for num in acc]
        row.extend(acc)
        df.loc[len(df)] = row 
    
    return df
```

```python
clfs = apply_pruning(ccp_alphas, x_train, y_train)
print("number of alpha values tested:", len(clfs))
```

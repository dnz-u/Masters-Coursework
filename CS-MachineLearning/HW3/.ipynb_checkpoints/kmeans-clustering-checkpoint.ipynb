```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import time
```

```python
def plot_img(img: np.array, centroids: np.array):
    q_image = quantize_image(img, centroids)
    plt.imshow(q_image, cmap=None)
    plt.axis('off')
    plt.title(f'K = {len(centroids)}')
    plt.tight_layout()
    plt.savefig(f'km{len(centroids)}.svg', format='svg', bbox_inches='tight')


def quantize_image(img: np.array, centroids: np.array) -> np.array:
    quantized_image = img.copy()
    labels = calculate_labels(img, centroids)
    dim0 = img.shape[0]
    dim1 = img.shape[1]
    
    # iterates over img pixels
    for i in range(dim0):
        for j in range(dim1):
            # (i,j) current pixel position
            centroid_index = int(labels[i][j])
            quantized_image[i][j] = centroids[centroid_index]
    
    return quantized_image
```

```python
def init_centroids(img, number_of_channels: int, K: int) -> np.array:

    n_img = img.reshape(-1, number_of_channels)

    centroids = list()
    centroid_indices = list()

    while len(centroids) < K:
        centroid = []

        centroid_index = np.random.choice(n_img.shape[0], size=1, replace=False).item()
        centroid = list(n_img[centroid_index])

        # append centroid if not exist
        if centroid not in centroids and centroid_index not in centroid_indices:
            centroid_indices.append(centroid_index)
            centroids.append(centroid)

    # returns it as a numpy array
    return np.array([np.array(centroid) for centroid in centroids])


def distance_2(pixel_vals, cluster):
    return np.sqrt(np.sum(np.power(pixel_vals - cluster, 2)))


def calculate_labels(img: np.array, centroids: np.array) -> np.array:
    # assign the closest label
    dim0 = img.shape[0]
    dim1 = img.shape[1]
    labels = np.zeros((dim0, dim1))
    
    # iterates over img pixels
    for i in range(dim0):
        for j in range(dim1):
            # (i,j) current pixel position
            
            # closest centroid distance search variables for a pixel
            best_distance = float('inf')
            best_distance_label = -1
            
            # closest centroid search
            for centroid_index, centroid in enumerate(centroids, start=0):
                pixel_vals = img[i, j, :]
                dist = distance_2(pixel_vals, centroid)
                
                if dist < best_distance:
                    best_distance = dist
                    best_distance_label = centroid_index
            
            # assign the closest label
            labels[i, j] = best_distance_label
    
    return labels


def update_centroids(img: np.array, labels: np.array, centroids: np.array) -> np.array:
    # assign the closest label
    dim0 = img.shape[0]
    dim1 = img.shape[1]
    
    c = img.shape[2]  # number of channel
    
    K = len(centroids)
    
    # occurence of each centroid(cluster)
    occ = np.zeros(K)
    total_vals = np.zeros((K, c))
    
    # iterates over img pixels
    for i in range(dim0):
        for j in range(dim1):
            # (i,j) current pixel position
            label = int(labels[i][j])
            occ[label] += 1
            total_vals[label] += img[i][j]
    
    # calculate mean
    new_centroids = np.zeros_like(centroids)
    for i in range(K):    
        # if no update leave the centroid as it is
        if occ[i] > 0:
            new_centroids[i] = total_vals[i] / occ[i]
        else:
            new_centroids[i] = centroids[i].copy()
    
    return new_centroids


def SSE_Loss(img: np.array, labels: np.array, centroids: np.array):
    dim0 = img.shape[0]
    dim1 = img.shape[1]
    
    sse_loss = 0
    
    # iterates over img pixels
    for i in range(dim0):
        for j in range(dim1):
            # (i,j) current pixel position
            label = int(labels[i][j])
            corresponding_centroid = centroids[label]
            pixel_vals = img[i][j]
            err = np.sum(np.power(pixel_vals - corresponding_centroid, 2))
        
            sse_loss += err
    
    return sse_loss/ (dim0*dim1)

def pp(pre, cur):
    return ((pre-cur)/pre)

def kmeans(img: np.array, K: int, max_epochs: int):
    # img x, y
    x, y = img.shape[0], img.shape[1]
    
    # img channels
    c = img.shape[2]
    
    # log data
    epoch_losses = list()
    epoch_centroids = list()
    
    # initialize K number of centroids
    centroids = init_centroids(img, number_of_channels=c, K=K)
    
    # update the log data for initialization
    labels = calculate_labels(img, centroids)
    sse_loss = SSE_Loss(img, labels, centroids)
    epoch_losses.append(sse_loss)
    epoch_centroids.append(centroids.copy())
    print(f"{K=} :")
    print(f"Initial SSE Loss: {sse_loss}") 
    
    for i in range(1, max_epochs+1):
        
        labels = calculate_labels(img, centroids)
        new_centroids = update_centroids(img, labels, centroids)
        
        sse_loss = SSE_Loss(img, labels, new_centroids)
                
        # it will append to index 1. index 0 will be initialization loss
        epoch_losses.append(sse_loss)
        epoch_centroids.append(new_centroids.copy())
        
        epsilon = 0.01  # 1 percent
        pre = epoch_losses[-2]
        cur = epoch_losses[-1]
        if pp(pre, cur) < epsilon:
            break       
        
        # continue to iterate with new centroids
        centroids = new_centroids.copy()
            
    print(f"Last SSE Loss: {sse_loss}\n") 

    return labels, centroids, epoch_losses, epoch_centroids
```

```python
k_cluster_size = list()
k_losses = list()
k_centroids = list()
k_times = list()

img = np.array(Image.open('sample.jpg'))/255

for k in [2, 3, 4, 5, 6]:
    start = time.perf_counter()

    labels, centroids, epoch_losses, epoch_centroids = kmeans(img, k, max_epochs=10)

    end = time.perf_counter()
    time_elapsed = end - start

    k_cluster_size.append(k)
    k_losses.append(epoch_losses[-1])
    k_centroids.append(epoch_centroids[-1].copy())
    k_times.append(time_elapsed)
    
    
```
